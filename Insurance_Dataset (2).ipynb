```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv('train.csv')
Data
```

```python
def Cleaning(X):
    id_=X['id']
    y_=X['Premium Amount']
    X_=X.drop(columns=['id','Premium Amount'],axis=1)
    X_['Timestamp']=pd.to_datetime(X_['Policy Start Date']).astype('int64')//10*9
    X_=X_.drop(columns=['Policy Start Date'],axis=1)
    for col in X_.select_dtypes(include=['float64', 'int64']).columns:
        X_[col].fillna(X_[col].median(), inplace=True)
    X_=X_.fillna('Unknown')
    bins=[0,15,25,35,45,55,60,100]
    labels=['<15','15-25','25-35','35-45','45-55','55-60','>60']
    bins1=[0,5,10,15,20,50]
    labels1=['<5','5-10','10-15','15-20','>20']
    X_['Age_group']=pd.cut(X_['Age'],bins=bins,labels=labels,right=False)
    X_['Vehicle_Age_group']=pd.cut(X_['Vehicle Age'],bins=bins1,labels=labels1,right=False)
    X_=X_.drop(columns=['Age','Vehicle Age'],axis=1)
    X_E=pd.get_dummies(X_,columns=['Age_group','Gender','Marital Status','Number of Dependents','Education Level','Occupation','Location','Policy Type','Previous Claims','Vehicle_Age_group','Insurance Duration','Customer Feedback','Smoking Status','Exercise Frequency','Property Type'])
    X_E_C=X_E.astype(int)
    X_E_C_N=((X_E_C-X_E_C.mean())/X_E_C.std())
    y_N=((y_-y_.mean())/y_.std())
    return X_E_C_N, id_,y_N,y_
```

```python
X,Id,y,X_cleaned=Cleaning(Data)
```

```python
def Linear_Regression(X,y,learning_rate,theta,b,iters):
    cost_history=np.zeros(iters)
    for i in range(iters):
        y_pred=X@theta+b
        #print(y_pred)
        gradient_theta=(np.dot(X.T,(y_pred-y)))/X.shape[1]
        gradient_b=((y_pred-y))/X.shape[1]
        theta=theta-(learning_rate*gradient_theta)
        b=b-(learning_rate*gradient_b)
        cost_history[i]=(1/(2*X.shape[1])) * np.sum((y_pred - y)**2)
    return theta, b, cost_history
```

```python

iters=10000
learning_rate=0.01
thetas=[]
cost_historys=[]
bs=[]
for i in range(1200):
    theta=np.zeros(70)
    b=0
    theta,b,cost_history=Linear_Regression(chunked_dfs[i].values,y1[i],learning_rate,theta,b,iters)
    thetas.append(theta)
    cost_historys.append(cost_history)
    bs.append(b)
    
```

```python
#itemwise_sum = np.sum(thetas, axis=0)
itemwise_avg1 = np.median(thetas, axis=0)

itemwise_avg3 = np.median(cost_historys, axis=0)
```

```python
import statistics
all_elements = [item for sublist in bs for item in sublist]

# Calculate median
b= statistics.median(all_elements)
```

```python
from sklearn.metrics import mean_squared_log_error
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))
```

```python
result = rmsle(y+10, y_pred+10)
print("RMSLE:", result)
```
