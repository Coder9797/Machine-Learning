```python
from sklearn.preprocessing import LabelEncoder
Data1=Data1.fillna('None')
Data1[['Systolic', 'Diastolic']] = Data1['Blood Pressure'].str.split('/', expand=True)
Data1.drop(columns=['Blood Pressure'],axis=1,inplace=True)
Data1=pd.get_dummies(Data1,columns=['Occupation','Gender','BMI Category'])
encoder = LabelEncoder()
Data1['Sleep_Disorder'] = encoder.fit_transform(Data1['Sleep Disorder'])
Data1.drop(columns='Sleep Disorder',axis=1,inplace=True)
Data1.astype(int)
Data1
```

```python
Data1.astype(int)
```

```python
y=Data1['Sleep_Disorder']
Data1.drop(columns=['Sleep_Disorder'],axis=1,inplace=True)
X=Data1
```

```python
X=X.astype(int)
```

```python
X
```

```python
y_train = pd.DataFrame(y, columns=['Y'])
y_train['Y']=y
#y_train = np.log1p(y_train)
```

```python
import lightgbm as lgb
```

```python
train_data = lgb.Dataset(X, label=y_train)
```

```python
params = {
    'objective': 'multiclass',     # Multi-class classification
    'metric': 'multi_logloss',     # Multi-class log loss metric
    'num_class': 3,                # Number of classes (3 in this example)
    'boosting_type': 'gbdt',       # Gradient Boosting Decision Trees
    'learning_rate': 0.05,
    'num_leaves': 31,
    'verbose': -1                  # Suppress output
}
```

```python
num_round = 500
model = lgb.train(params, train_data, num_round)
```

```python
y_pred = model.predict(X, num_iteration=model.best_iteration)
```

```python
#y_pred = np.expm1(y_pred)
```

```python
y_pred_labels = y_pred.argmax(axis=1)
```

```python
y_pred_labels
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
cm = confusion_matrix(y, y_pred_labels)

# Plot the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for 3-Class Classification')
plt.show()
```

```python
from sklearn.metrics import balanced_accuracy_score

balanced_acc = balanced_accuracy_score(y, y_pred_labels)
print("Balanced Accuracy:", balanced_acc)
```
