```python
X_Standardized=(X-X.mean())/X.std()
X_Standardized.shape
```

```python
array_data=X_Standardized.values
hypothesis=np.random.rand(41,1)
y=y.values
y_standardized=(y-y.mean())/y.std()
y_actual = y_standardized.reshape(-1, 1)
y_actual
```

```python
def gradient_descent(array_data,hypothesis,y_actual,learning_rate,iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        y_prec=array_data.dot(hypothesis)
        gradients = (1 / 41) * np.dot(array_data.T, (y_prec - y_actual))
        hypothesis=hypothesis-(learning_rate*gradients)
        cost_history[i]=(1/(2*41)) * np.sum((y_prec - y_actual)**2)
    return hypothesis,cost_history
```

```python
iterations=3500
learning_rate=0.003
hypothesis,cost_history=gradient_descent(array_data,hypothesis,y_actual,learning_rate,iterations)
hypothesis
cost_history
```

```python
import matplotlib.pyplot as plt

plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Over Iterations")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
y_prec=array_data.dot(hypothesis)
r_squared = r2_score(y_actual, y_prec)  # Use r2_score to calculate R-squared

print(f"R-squared: {r_squared*100}")
```

```python
n = len(y_prec)
k = array_data.shape[1]

adjusted_r2 = 1 - (1 - r_squared) * (n - 1) / (n - k - 1)


print(f"Adjusted R²: {adjusted_r2*100}")
```
